```python
#here we will scrap some webpages using read_html function which generally extracts tables contents form html
#drawback : pd.read_html will not read all contents of an HTML table that are not seen due to pagination or dynamic sorting/filtering. It only processes the HTML source code it receives at the time of the request.
import pandas as pd
simpsons = pd.read_html("https://en.wikipedia.org/wiki/List_of_The_Simpsons_episodes_(seasons_1%E2%80%9320)") #this function extracts all the tables from html page
len(simpsons)
```

```python
simpsons[1]
```

```python
from google.colab import files

# Install xlsxwriter
!pip install xlsxwriter

output_filename = "Simpsons_Episodes.xlsx"

with pd.ExcelWriter(output_filename, engine='xlsxwriter') as writer:
    for i, df in enumerate(simpsons):
        # Check if columns are MultiIndex
        if isinstance(df.columns, pd.MultiIndex):
            # Flatten the MultiIndex columns
            # This creates a more descriptive single-level column name
            df.columns = ['_'.join(col).strip() if isinstance(col, tuple) else col for col in df.columns.values]
            # You might want to clean up "_nan" or "nan_" if they appear
            df.columns = [col.replace('nan_', '').replace('_nan', '') for col in df.columns]

        sheet_name = f"Table_{i+1}"
        df.to_excel(writer, sheet_name=sheet_name, index=False) # Keep index=False

print(f"All tables exported to {output_filename} in your Google Colab environment.")
files.download(output_filename)
print(f"The file '{output_filename}' has been downloaded to your local machine.")
```
